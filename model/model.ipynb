```python
# normalize the data
from sklearn.preprocessing import StandardScaler
scalar= StandardScaler()
X_scaled= scalar.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)
```

```python
# logistic reg model
def sigmoid(z):
    g= 1/ (1+ np.exp(-z))
    return g 
```

```python
def cost_func(x,y,w,b):
    m, n= x.shape
    total_cost= 0.0
    for i in range(m):
        f_wb = sigmoid(np.dot(x[i], w) + b )
        total_cost = total_cost - y[i] * ( np.log(f_wb) ) - (1- y[i])* (np.log(1- f_wb))
    total_cost= total_cost/m  
    total_cost= float(total_cost)

    return total_cost
```

```python
def compute_gradient(x,y,w,b):
    m,n= x.shape
    dj_dw = np.zeros(w.shape)
    dj_db = 0.0
    for i in range(m):
        f_wb= sigmoid(np.dot(x[i], w) + b)
        err = (f_wb - y[i])
        dj_db += err
        for j in range(n):
            dj_dw[j] += err * x[i,j]
    dj_db= dj_db/m
    dj_dw= dj_dw/m
    return dj_db, dj_dw
```

```python
import math

def gradient_descent(X, y, w_in, b_in, cost_function, gradient_function, alpha, num_iters): 
    m = len(X)
    
    J_history = []
    w_history = []
    
    for i in range(num_iters):
        dj_db, dj_dw = gradient_function(X, y, w_in, b_in)   

        # Update Parameters using w, b, alpha and gradient
        w_in = w_in - alpha * dj_dw               
        b_in = b_in - alpha * dj_db              
       
        # Save cost J at each iteration
        if i<100000:      # prevent resource exhaustion 
            cost =  cost_function(X, y, w_in, b_in)
            J_history.append(cost)

        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(num_iters/10) == 0 or i == (num_iters-1):
            w_history.append(w_in)
            print(f"Iteration {i:4}: Cost {float(J_history[-1]):8.2f}   ")
        
    return w_in, b_in, J_history, w_history #return w and J,w history for graphing
```

```python
def predict(x,w,b):
    m, n = x.shape
    p = np.zeros(m)
    for i in range(m):
        z= (np.dot(x[i], w) + b)
        f_wb= sigmoid(z)
        p[i] = 1 if f_wb>=0.5 else 0
    return p

```

```python

initial_w = 0.01 * (np.random.rand(30) - 0.5)
initial_b = 0.0

# Some gradient descent settings
iterations = 10000
alpha = 0.001

w,b, J_history,_ = gradient_descent(X_train ,y_train, initial_w, initial_b, 
                                   cost_func, compute_gradient, alpha, iterations)
```

#### Predicting on test data

```python
y_predict = predict(X_test, w,b).astype(int)
```

```python
class CustomLogisticModel:
    def __init__(self, w, b, mean=None, std=None):
        self.w = w
        self.b = b
        self.mean = mean
        self.std = std

    def scale(self, X):
        if self.mean is not None and self.std is not None:
            return (X - self.mean) / self.std
        return X

    def predict(self, X):
        X = self.scale(X)
        z = np.dot(X, self.w) + self.b
        return (1 / (1 + np.exp(-z)) >= 0.5).astype(int)
```

```python
model = CustomLogisticModel(w, b, X_train.mean(axis=0), X_train.std(axis=0))

import pickle
with open("model.pkl", "wb") as f:
    pickle.dump(model, f)
with open("scalar.pkl", 'wb') as f:
    pickle.dump(scalar, f)
```
